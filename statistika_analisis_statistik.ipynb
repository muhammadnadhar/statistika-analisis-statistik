```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

Analitis statistika

```python
# Load dataset CSV
file_path = "/content/survey_results_public.csv"
df = pd.read_csv(file_path)

# Menampilkan 5 baris pertama
#df.head()



# Cek nama kolom yang relevan
#print(df.columns)

# Filter data pengguna Next.js
df_nextjs = df[df['WebframeHaveWorkedWith'].str.contains("Next.js", na=False)]

# Menampilkan beberapa baris pertama
df_nextjs.head()



# Statistik deskriptif pengalaman pengguna Next.js
#print(df_nextjs["YearsCodePro"].describe())

# Visualisasi distribusi pengalaman pengguna Next.js
plt.figure(figsize=(8, 5))
sns.histplot(df_nextjs["YearsCodePro"].dropna(), bins=10, kde=True)
plt.title("Distribusi Pengalaman Developer Next.js")
plt.xlabel("Tahun Pengalaman")
plt.ylabel("Frekuensi")
plt.show()



# URL GitHub API untuk repo Next.js
url = "https://api.github.com/repos/vercel/next.js"

# Jika menggunakan token, tambahkan ke headers
# headers = {"Authorization": "token YOUR_GITHUB_TOKEN"}
response = requests.get(url)
data = response.json()

# Menampilkan data penting
print(f"Next.js Stars: {data['stargazers_count']}")
print(f"Forks: {data['forks_count']}")
print(f"Watchers: {data['watchers_count']}")



# Simulasi data tren Next.js berdasarkan Google Trends atau GitHub Stars
trend_data = {
    "Year": [2018, 2019, 2020, 2021, 2022, 2023],
    "Popularity": [40, 55, 70, 85, 90, 95]  # Skala 0-100
}

df_trends = pd.DataFrame(trend_data)

# Visualisasi tren popularitas Next.js
plt.figure(figsize=(8,5))
sns.lineplot(x="Year", y="Popularity", data=df_trends, marker="o")
plt.title("Tren Popularitas Next.js (2018-2023)")
plt.xlabel("Tahun")
plt.ylabel("Popularitas")
plt.grid()
plt.show()

```

Versi yang di tingkatkan dan di animasikan

```python
#  Load dataset CSV
file_path = "/content/survey_results_public.csv"
df = pd.read_csv(file_path)

#  Filter data pengguna Next.js
df_nextjs = df[df['WebframeHaveWorkedWith'].str.contains("Next.js", na=False)]

#  Statistik deskriptif pengalaman pengguna Next.js
print(" Statistik Pengalaman Developer Next.js:\n")
print(df_nextjs["YearsCodePro"].describe())

#  Visualisasi distribusi pengalaman developer Next.js
plt.figure(figsize=(8, 5))
sns.histplot(df_nextjs["YearsCodePro"].dropna(), bins=10, kde=True, color="blue")
plt.title("Distribusi Pengalaman Developer Next.js", fontsize=14, fontweight="bold")
plt.xlabel("Tahun Pengalaman", fontsize=12)
plt.ylabel("Frekuensi", fontsize=12)
plt.grid(alpha=0.3)
plt.show()

#  URL GitHub API untuk repo Next.js
url = "https://api.github.com/repos/vercel/next.js"

#  Jika menggunakan token, tambahkan ke headers
# headers = {"Authorization": "token YOUR_GITHUB_TOKEN"}
response = requests.get(url)
data = response.json()

#  Menampilkan data penting dari GitHub
print("\n Statistik GitHub Next.js:\n")
print(f" Stars: {data['stargazers_count']}")
print(f" Forks: {data['forks_count']}")
print(f" Watchers: {data['watchers_count']}")

#  Simulasi data tren Next.js berdasarkan Google Trends atau GitHub Stars
trend_data = {
    "Year": [2018, 2019, 2020, 2021, 2022, 2023],
    "Popularity": [40, 55, 70, 85, 90, 95]  # Skala 0-100
}
df_trends = pd.DataFrame(trend_data)

#  Animasi Tren Popularitas Next.js
plt.figure(figsize=(8, 5))
plt.title("Tren Popularitas Next.js (2018-2023)", fontsize=14, fontweight="bold")
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Popularitas", fontsize=12)
plt.grid(alpha=0.3)

for i in range(1, len(df_trends) + 1):
    plt.plot(df_trends["Year"][:i], df_trends["Popularity"][:i], marker="o", linestyle="-", color="red")
    plt.pause(0.5)  #  Animasi naik-turun setiap 0.5 detik

plt.show()
```
